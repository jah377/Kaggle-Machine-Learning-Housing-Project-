#### Continuous Numerical Predictors EDA

```python
# Plot sales price as a function of CONTINUOUS numerical variable data
ncol = 4  # number of columns

plt.figure(figsize=(15, 20))
for idx, col in enumerate(num_cont_features):
    plt.subplot(math.ceil(len(num_cont_features) / ncol), ncol, idx + 1)
    plt.scatter(x=house_data[col], y=house_data["SalePrice_log"])
    plt.ylabel("SalePrice_log")
    plt.xlabel(col)
```

```python
# Determine Outliers of Continuous Data
def outlier_idx(data, thresh):

    avg = data.mean()  # calc average
    stdev = data.std()  # calc standard deviation
    z_score = (data - avg) / stdev  # calc z_score
    ol = z_score > thresh  # boolean (True=outlier)
    outlier_idx = data.index[ol]  # indexes of outliers

    return list(outlier_idx)


outlier_store = {}
thresh = 6  # outlier > +6std away from mean

for idx, col in enumerate(num_cont_features):
    outliers = outlier_idx(house_data[col], thresh)  # outlier indices

    if len(outliers) > 0:
        outlier_store[col] = outliers  # only store features with outliers
```

```python
# Plot Outliers
plt.figure(figsize=(15, 20))
for idx, col in enumerate(num_cont_features):
    outliers = outlier_idx(house_data[col], 6)  # outlier indices
    plt.subplot(math.ceil(len(num_cont_features) / ncol), ncol, idx + 1)
    plt.scatter(x=house_data[col], y=house_data["SalePrice_log"], c="blue")
    plt.scatter(
        x=house_data[col][outliers], y=house_data["SalePrice_log"][outliers], c="red"
    )
    plt.ylabel("SalePrice_log")
    plt.xlabel(col)
```

```python
# Function to Quickly Summarize NA, Zero, and Correlation statistics
def dirty_summary(features):
    store = [["Predictor", "n_NA", "perc_NA", "n_zero", "perc_zero", "corr"]]

    for idx, feature in enumerate(features):
        data = house_data[[feature, "SalePrice_log"]]  # create df subset
        n_missing = data[feature].isna().sum()  # calc miss values
        perc_missing = n_missing / data[feature].shape[0] * 100  # calc %miss values

        new_data = data.dropna()  # remove missing data
        n_zero = sum(new_data[feature] == 0)  # calc num of zeros
        perc_zero = n_zero / new_data.shape[0] * 100  # calc %zeros
        corr_dropna = stats.pearsonr(
            new_data[feature].values, new_data["SalePrice_log"].values
        )[0]

        store.append(
            [
                feature,
                n_missing,
                round(perc_missing, 1),
                n_zero,
                round(perc_zero, 1),
                round(corr_dropna, 2),
            ]
        )

    analyses = pd.DataFrame(store[1:], columns=store[0])
    return analyses
```

```python
continuous_summary = dirty_summary(num_cont_features)
continuous_summary
```

```python
from sklearn.impute import SimpleImputer

imp_vars = continuous_summary["Predictor"][continuous_summary.n_NA > 0]
feature = imp_vars[0]

for x in imp_vars:
    data = pd.concat([house_data[x], house_data["SalePrice_log"]], axis=1)
    imp_mean = SimpleImputer(missing_values=np.nan, strategy="mean")
    imp_mean.fit(data)
    mean_imp = imp_mean.transform(data)

    imp_med = SimpleImputer(missing_values=np.nan, strategy="median")
    imp_med.fit(data)
    med_imp = imp_med.transform(data)

    test = pd.concat(
        [data.iloc[:, 0], pd.DataFrame(mean_imp[:, 0]), pd.DataFrame(med_imp[:, 0])],
        axis=1,
    )
    test.columns = ["missing", "mean_imp", "med_imp"]

    plt.figure()
    for i in test.columns:
        sns.distplot(test[i], hist=False, kde=True, label=i)
        plt.title(x)
```
