#### Impute Lot Frontage

```python
# Impute by cumulative mean value
mean_imp = df["LotFrontage"].fillna(df["LotFrontage"].mean())

# Impute by mode value
vals = df["LotFrontage"].loc[df["LotFrontage"].notna()].tolist()
lf_mode = max(set(vals), key=vals.count)
mode_imp = df["LotFrontage"].fillna(lf_mode)


# Impute by mean per neighborhood
bynbhood_imp = df.groupby("Neighborhood")["LotFrontage"].transform(
    lambda x: x.fillna(x.median())
)

# Plot density plot of various imputations methods
# format data for plot
imputation_df = pd.DataFrame(
    {
        "data": df.LotFrontage.loc[df.LotFrontage.notna()],
        "by_neighborhood": bynbhood_imp,
        "mean": mean_imp,
        "mode": mode_imp,
    }
)


# Plot all imputation methods
plt.figure(figsize=(10, 6))
for method in imputation_df.columns:

    color_dict = {
        "data": "red",
        "by_neighborhood": "green",
        "mean": "grey",
        "mode": "grey",
    }

    if method in ["data", "by_neighborhood"]:
        kde_kws = {"linewidth": 5}
    else:
        kde_kws = {"linewidth": 2}

    sns.distplot(
        imputation_df[method],
        hist=False,
        kde=True,
        kde_kws=kde_kws,
        label=method,
        color=color_dict[method],
    )
sns.set(style="whitegrid", font_scale=1.5)
plt.title("Density Plot By Imputation Method")
plt.xlabel("LotFrontage (ft^2)")
plt.ylabel("Density")

if save_figures:
    plt.savefig("saved_figures/imputation_example.tiff", bbox_inches="tight")
# ==============================================================
# Based on the figure, imputation by neighborhood mean is most accurate method
df["LotFrontage"] = bynbhood_imp
```

## Remove Outliers
2D PCA used to identify outliers based on features with continuous numerical data

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
from preprocess_module import outlier_idx

#Identify features with highest variability (based on unique values)
top_cont_feat = (
    df.select_dtypes(np.number)
    .nunique()
    .sort_values(ascending=False)
    .index.tolist()[:15]
)
continous_features = df[top_cont_feat]
normalized_data = preprocessing.normalize(continous_features) #normalize data

#Perform PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(normalized_data)
pca_df = pd.DataFrame(data = principalComponents, columns = ['PC1', 'PC2'])

#Identify observations >6 standard deviations
outliers = outlier_idx(pca_df.PC2, 6) + outlier_idx(pca_df.PC1, 5)
print('Number of identified outliers: {0}'.format(len(outliers)))

# Format dataframe for plotting data and outliers
label = pd.Series(["data"] * pca_df.shape[0], name="Label")
pca_df = pd.merge(pca_df, label, left_index=True, right_index=True)
pca_df.Label.iloc[outliers] = 'outlier'

# Plot data identifying outliers
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df,
                x='PC1', 
                y='PC2', 
                hue='Label',
                x_jitter=True,
                y_jitter=False)
sns.set(style="whitegrid", font_scale=1.5)
plt.title("Outlier Identification using PCA")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
if save_figures: plt.savefig("saved_figures/pca_outliers.tiff", bbox_inches="tight")

#Remove outliers
df.drop(outliers, axis=0, inplace=True)
```

#### Continuous Features

```python
# Age = Year sold - year built
df["Age"] = df["YrSold"] - df["YearBuilt"]

# PreWWII = all houses built before 1945
df["PreWWII"] = np.where(df["YearBuilt"] < 1945, 1, 0).tolist()

# AgeRemod = Year sold - year remodeled
df["AgeRemod"] = df["YrSold"] - df["YearRemodAdd"]

# HasHouseRemod = 0 (no) or 1 (yes)
df["HasHouseRemod"] = np.where(df["YearBuilt"] - df["YearRemodAdd"] > 0, 1, 0)

# HasGarageRemod = 0 (no) or 1 (yes)
df["HasGarageRemod"] = np.where(df["GarageYrBlt"] - df["YearBuilt"] > 0, 1, 0)
df.drop('GarageYrBlt', axis=1, inplace=True)

# Total area of house = Above ground living area + basement living area
df["TotalSF"] = df["GrLivArea"] + df["TotalBsmtSF"]

# Total porch SF = OpenPorchSF + EnclosedSF + 3SsnPorch + ScreenPorch
df["TotPorchSF"] = (
    df["OpenPorchSF"] + df["EnclosedPorch"] + df["3SsnPorch"] + df["ScreenPorch"]
)

# Total bathrooms = Full bath + 0.5*Halfbath (for both house and basement)
df["TotBaths"] = (
    df["FullBath"]
    + df["BsmtFullBath"]
    + 0.5 * df["HalfBath"]
    + 0.5 * df["BsmtHalfBath"]
)

#Has Pool = 0 (no) or 1 (yes)
df['HasPool'] = np.where(df["PoolArea"] > 0, 1, 0)
```

#### Categorical Features

```python
from preprocess_module import explore_feat

#Use function to explore count and mean SalePrice per categorical feature
explore_feat(df, "Condition1", sortby="mean")
```

```python
# MSZoning: Condence FV, RH, and C into other (only 6%)
df['MSZoning'] = df['MSZoning'].map({'RL': 'RL', 'RM':'RM'})
df['MSZoning'].fillna('other', inplace=True) 

# LotShape: Combine moderately and severely irregular lotshapes
df['LotShape'] = df['LotShape'].map({'Reg': 'Reg', 'IR1':'IR1'})
df['LotShape'].fillna('other', inplace=True) 

# LotConfig: Combine properties with frontage (FR2 and FR3)
df['LotConfig'] = df['LotConfig'].map({'Inside': 'Inside', 'Corner':'Corner',
                                     'CulDSac':'CulDSac'})
df['LotConfig'].fillna('HasFrontage', inplace=True) 

# Condition1/2: Combine positive features (PosN, PosA) and negative features (RRNe, RRNn)
dict_condition = {
    "Artery": "Artery",
    "Feedr": "Artery",
    "Norm": "Norm",
    "PosA": "Pos",
    "PosN": "Pos",
    "RRAe": "RRAe",
    "RRAn": "RRAn",
    "RRNe": "AdjRR_200ft",
    "RRNn": "AdjRR_200ft"}
df['Condition1'] = df['Condition1'].map(dict_condition)
df['Condition2'] = df['Condition2'].map(dict_condition)

# Exterior1st: Condense BrkComm, Stone, AsphShn, CBlock, ImStucc to other
df['Exterior1st'].replace(['Brk Cmn', 'BrkComm', 'Stone', 'AsphShn', 'CBlock', 'ImStucc'],'Other', inplace=True)
df['Exterior2nd'].replace(['Brk Cmn', 'BrkCmn', 'Stone', 'AsphShn', 'CBlock', 'ImStucc'],'Other', inplace=True)

# Foundation: Condense Wood, Stone, and Slab to other
df['Foundation'].replace(['Slab', 'Stone', 'Wood'],'other', inplace=True)

# RoofStyle: Combine Hip and Shed style
df['RoofStyle'].replace('Shed','Hip', inplace=True)

# MiscFeature: convert to binary ( 1=yes, 0=no)
df['HasMiscFeat'] = np.where( df['MiscFeature'] == 'No MiscFeature', 0, 1)
df.drop('MiscFeature', axis=1, inplace=True)

# Electrical: convert to binary ( 1=yes, 0=no)
df['HasModernCircBreak'] = np.where( df['Electrical'] == 'SBrkr', 1, 0)
df.drop('Electrical', axis=1, inplace=True)

# Heating: Convert to boolean (1=yes, 0=no)
df['HasGasHeating'] = np.where( (df.Heating == 'GasA') | (df.Heating == 'GasW'), 1, 0)
df.drop('Heating', axis=1, inplace=True)

# RoofStyle: Convert to boolean (1=yes, 0=no)
df['HasCompRoofMatl'] = np.where(df['RoofMatl']=='CompShg', 1, 0)
df.drop('RoofMatl', axis=1, inplace=True)

# LandSlope: Convert to binary (1=yes, 0=no)
df['HasLandSlope'] = np.where(df["LandSlope"] == "Gtl", 0, 1)
df.drop('LandSlope', axis=1, inplace=True)

# Utilities: convert to boolean (1=yes, 0=no)
df['HasPubUtil'] = np.where(df["Utilities"] == "AllPub", 1, 0)
df.drop('Utilities', axis=1, inplace=True)

# LotShape: convert to boolean (1=yes, 0=no)
df['HasFlatLand'] = np.where(df["LandContour"] == "Lvl", 1, 0)
df.drop('LandContour', axis=1, inplace=True)

# Street: convert to boolean (1=yes, 0=no)
df["HasPavedStrt"] = np.where(df["Street"] == "Pave", 1, 0)
df.drop('Street', axis=1, inplace=True)

# CentralAir: convert to boolen (1 = yes; 0 = No)
df["HasCentralAir"] = np.where(df["CentralAir"] == "Y", 1, 0)
df.drop('CentralAir', axis=1, inplace=True)

# Convert Alley to boolean (1=yes, 0=no)
df["HasAlley"] = np.where(df["Alley"] == "No Alley", 0, 1)
df.drop('Alley', axis=1, inplace=True)
```
